```python
import cv2

def play_video_in_reverse_slow_motion(input_video_path, output_video_path, slow_motion_factor=0.5):
    """
    Plays a given video in reverse and slow motion.

    Args:
        input_video_path (str): The path to the input video file.
        output_video_path (str): The path where the processed video will be saved.
        slow_motion_factor (float): A factor to slow down the video.
                                    E.g., 0.5 means half the original speed,
                                    2.0 means twice the original speed (faster).
                                    Must be greater than 0. Ideally, between 0 and 1 for slow motion.
    """
    cap = cv2.VideoCapture(input_video_path)

    if not cap.isOpened():
        print(f"Error: Could not open video file {input_video_path}")
        return

    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Calculate new FPS for slow motion
    new_fps = fps * slow_motion_factor

    # Define the codec and create VideoWriter object
    # Use 'mp4v' for .mp4 files. Other codecs might be 'XVID', 'MJPG', etc.
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, new_fps, (frame_width, frame_height))

    if not out.isOpened():
        print(f"Error: Could not create output video file {output_video_path}")
        cap.release()
        return

    # Read all frames into a list
    frames = []
    for _ in range(total_frames):
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)

    # Write frames in reverse order
    for frame in reversed(frames):
        out.write(frame)

    print(f"Video successfully processed and saved to {output_video_path}")

    cap.release()
    out.release()
    cv2.destroyAllWindows()

# Example Usage (uncomment to run, make sure to replace with your video paths):
# input_video = 'your_input_video.mp4'
# output_video = 'reversed_slow_motion_video.mp4'
# play_video_in_reverse_slow_motion(input_video, output_video, slow_motion_factor=0.5)
```
