```python
import cv2
import numpy as np
from IPython.display import Video, display
import os
```

```python
def reverse_video(input_video_path, output_video_path):
    """
    Reverses the frames of a video to create a video in reverse mode.

    Args:
        input_video_path (str): The path to the input video file.
        output_video_path (str): The path to save the reversed video file.
    """
    # Create a dummy video file for demonstration if it doesn't exist
    if not os.path.exists(input_video_path):
        print(f"Creating a dummy video file at {input_video_path} for demonstration...")
        fourcc = cv2.VideoWriter_fourcc(*'MJPG') # Codec
        out = cv2.VideoWriter(input_video_path, fourcc, 20.0, (640, 480))

        for i in range(100):
            img = np.zeros((480, 640, 3), dtype=np.uint8)
            cv2.putText(img, f'Frame {i}', (50, 240), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 3, cv2.LINE_AA)
            out.write(img)
        out.release()
        print("Dummy video created.")

    # Open the input video file
    cap = cv2.VideoCapture(input_video_path)

    if not cap.isOpened():
        print(f"Error: Could not open input video file {input_video_path}")
        return

    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"Input video properties: {frame_width}x{frame_height}, {fps} FPS, {total_frames} frames")

    # Define the codec and create VideoWriter object
    # Use 'mp4v' for MP4 output, 'XVID' or 'MJPG' for AVI
    # The choice of codec can depend on the system and installed codecs
    fourcc = cv2.VideoWriter_fourcc(*'mp4v') # Codec for .mp4 files
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

    if not out.isOpened():
        print(f"Error: Could not open output video file {output_video_path}")
        cap.release()
        return

    frames = []
    # Read all frames into a list
    for i in range(total_frames):
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)

    # Write frames in reverse order to the output video
    print(f"Writing {len(frames)} frames in reverse to {output_video_path}...")
    for frame in reversed(frames):
        out.write(frame)

    # Release everything if job is finished
    cap.release()
    out.release()
    print(f"Video successfully reversed and saved to {output_video_path}")


# Example usage:
input_file = "video 1.jpg"
output_file = "reversed_video.mp4"

reverse_video(input_file, output_file)

# Display the reversed video if it was created successfully
if os.path.exists(output_file):
    print("\nDisplaying the reversed video:")
    display(Video(output_file, embed=True))
else:
    print(f"Could not find the output video file: {output_file}")
```
