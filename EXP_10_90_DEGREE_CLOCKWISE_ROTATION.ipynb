### Image Rotation

This task involves performing a 90-degree clockwise rotation on an image. We'll use the `OpenCV` library for image manipulation and `matplotlib` for displaying the images.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow

# Function to display images
def display_images(img1, title1, img2, title2):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
    plt.title(title1)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
    plt.title(title2)
    plt.axis('off')
    plt.show()
```

```python
# Load a sample image
# You can replace this URL with the path to your own image file.
image_path = '/content/image 2.jpg'

# Load the image from the local path
original_image = cv2.imread(image_path)


if original_image is None:
    print(f"Error: Could not load image from {image_path}")
else:
    print("Original image loaded successfully.")
    # Perform 90-degree clockwise rotation
    rotated_image = cv2.rotate(original_image, cv2.ROTATE_90_CLOCKWISE)

    # Display original and rotated images
    display_images(original_image, 'Original Image', rotated_image, 'Rotated 90 Degrees Clockwise')
```
