### Image Scaling Techniques

This demonstration will use the `Pillow` library (PIL) to perform image scaling. We will:
1. Create a simple dummy image.
2. Resize the image to a larger size (upscaling).
3. Resize the image to a smaller size (downscaling).
4. Display the original and resized images.

```python
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import numpy as np

# 1. Load the specified image for demonstration
image_path = '/content/images.nikki.jpg'
original_image = Image.open(image_path).convert('RGB') # Ensure it's RGB
original_size = original_image.size

print(f"Original image size: {original_image.size}")

# Display the original image
plt.figure(figsize=(3,3))
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# 2. Resize the image to a larger size (upscaling)
# Let's upscale to 200x200
larger_size = (200, 200)
upscaled_image = original_image.resize(larger_size, Image.Resampling.LANCZOS) # LANCZOS for high-quality downsampling/upsampling

print(f"Upscaled image size: {upscaled_image.size}")

# Display the upscaled image
plt.figure(figsize=(3,3))
plt.imshow(upscaled_image)
plt.title('Upscaled Image')
plt.axis('off')
plt.show()
```

```python
# 3. Resize the image to a smaller size (downscaling)
# Let's downscale to 50x50
smaller_size = (50, 50)
downscaled_image = original_image.resize(smaller_size, Image.Resampling.LANCZOS) # LANCZOS for high-quality downsampling/upsampling

print(f"Downscaled image size: {downscaled_image.size}")

# Display the downscaled image
plt.figure(figsize=(3,3))
plt.imshow(downscaled_image)
plt.title('Downscaled Image')
plt.axis('off')
plt.show()
```

### Comparison of Original and Resized Images

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(original_image)
plt.title(f'Original ({original_image.size[0]}x{original_image.size[1]})')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(upscaled_image)
plt.title(f'Upscaled ({upscaled_image.size[0]}x{upscaled_image.size[1]})')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(downscaled_image)
plt.title(f'Downscaled ({downscaled_image.size[0]}x{downscaled_image.size[1]})')
plt.axis('off')

plt.tight_layout()
plt.show()
```
