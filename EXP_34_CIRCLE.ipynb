```python
import cv2
import numpy as np

def create_image_with_circle(width, height, center_x, center_y, radius, color=(0, 0, 0), thickness=2):
    """
    Creates a white image of specified size and draws a circle on it.

    Args:
        width (int): The width of the image.
        height (int): The height of the image.
        center_x (int): The x-coordinate of the circle's center.
        center_y (int): The y-coordinate of the circle's center.
        radius (int): The radius of the circle.
        color (tuple): The color of the circle in BGR format (default is black).
        thickness (int): The thickness of the circle border. Use -1 for a filled circle.

    Returns:
        numpy.ndarray: The image with the drawn circle.
    """
    # Create a white image (height, width, 3 channels for BGR)
    # np.ones creates an array filled with ones, multiply by 255 for white pixels
    image = np.full((height, width, 3), 255, dtype=np.uint8)

    # Draw the circle on the image
    cv2.circle(image, (center_x, center_y), radius, color, thickness)

    return image
```

### Example Usage:

Let's create a 500x500 pixel white image with a red circle in the center.

```python
# Define image dimensions
img_width = 500
img_height = 500

# Define circle parameters
circle_center_x = img_width // 2
circle_center_y = img_height // 2
circle_radius = 100
circle_color = (0, 0, 255)  # Red color in BGR format
circle_thickness = -1  # -1 for a filled circle

# Create the image with the circle
result_image = create_image_with_circle(img_width, img_height, circle_center_x, circle_center_y, circle_radius, circle_color, circle_thickness)

# Display the image (requires an environment where cv2.imshow can open a window)
# If running in Colab, cv2.imshow might not work directly. Consider saving the image or using matplotlib.
# For demonstration, we'll save it to a file.
output_filename = 'image_with_circle.png'
cv2.imwrite(output_filename, result_image)

print(f"Image saved as '{output_filename}'")

# To display in Colab, you can use:
from google.colab.patches import cv2_imshow
cv2_imshow(result_image)
```
