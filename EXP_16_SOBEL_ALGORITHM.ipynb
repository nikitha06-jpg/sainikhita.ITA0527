```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
# Replace 'path/to/your/image.jpg' with the actual path to your image file
image_path = 'image 3.jpg'
img = cv2.imread(image_path)

# Check if the image was loaded successfully
if img is None:
    print(f"Error: Could not load image from {image_path}")
else:
    # Convert the image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply Sobel filter in X direction
    sobelx = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=3)

    # Apply Sobel filter in Y direction
    sobely = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=3)

    # Compute the magnitude of the gradient
    magnitude = np.sqrt(sobelx**2 + sobely**2)

    # Normalize the magnitude image to 0-255 for display
    magnitude = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)

    # Display the original and filtered images
    print("Original Image:")
    cv2_imshow(img)
    print("Sobel Filtered Image (Magnitude):")
    cv2_imshow(magnitude)
```

### Explanation of the Sobel Filter Code:

1.  **Import Libraries**: `cv2` for OpenCV functions, `numpy` for numerical operations, and `cv2_imshow` for displaying images in Colab.
2.  **Load Image**: Reads the input image from the specified path. It checks if the image loaded successfully.
3.  **Convert to Grayscale**: The Sobel filter is typically applied to grayscale images, so the color image is converted using `cv2.cvtColor()`.
4.  **Apply Sobel Filters**:
    *   `cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=3)` computes the gradient in the X direction. `cv2.CV_64F` specifies the output image depth (64-bit floating-point) to avoid truncation of derivatives.
    *   `cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=3)` computes the gradient in the Y direction. `ksize=3` specifies the size of the Sobel kernel (3x3).
5.  **Compute Magnitude**: The `magnitude` of the gradient is calculated as the square root of the sum of the squares of the X and Y gradients. This gives an overall edge strength.
6.  **Normalize for Display**: The `magnitude` image, which is in `CV_64F` format, is normalized to an 8-bit unsigned integer format (`CV_8U`) and scaled to the range 0-255 for proper display.
7.  **Display Images**: The original and the Sobel-filtered (magnitude) images are displayed using `cv2_imshow()`.
