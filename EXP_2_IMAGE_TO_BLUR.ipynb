```python
import cv2
import matplotlib.pyplot as plt

# Replace 'path/to/your/image.jpg' with the actual path to your image file
image_path = '/content/images.nikki.jpg'

# Read the image
# cv2.imread returns a NumPy array
# If the image path is incorrect or the image is not found, img will be None
img = cv2.imread(image_path)

# Check if the image was loaded successfully
if img is None:
    print(f"Error: Could not load image from {image_path}")
else:
    # Convert the image from BGR to RGB for matplotlib display
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Apply Gaussian Blur
    # The kernel size must be a positive, odd integer tuple (width, height)
    # The standard deviation in X and Y direction (sigmaX, sigmaY) can be 0,
    # in which case they are calculated from the kernel size.
    blurred_img = cv2.GaussianBlur(img_rgb, (51, 51), 0)

    # Display the original and blurred images side-by-side
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(blurred_img)
    plt.title('Gaussian Blurred Image')
    plt.axis('off')

    plt.show()
```
