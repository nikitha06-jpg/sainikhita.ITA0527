```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def add_text_to_image(image_path, text_to_add, output_path='image_with_text.jpg',
                        org=(50, 50), font=cv2.FONT_HERSHEY_SIMPLEX, font_scale=1,
                        color=(255, 255, 255), thickness=2):
    """
    Adds a given text string to an image using OpenCV.

    Args:
        image_path (str): The path to the input image.
        text_to_add (str): The text string to be added to the image.
        output_path (str): The path to save the output image with text.
        org (tuple): The bottom-left corner of the text string in the image (x, y).
        font (int): Font type, e.g., cv2.FONT_HERSHEY_SIMPLEX.
        font_scale (float): Font scale factor that is multiplied by the font-specific base size.
        color (tuple): Text color (B, G, R) for 8-bit images.
        thickness (int): Thickness of the lines used to draw the text.

    Returns:
        bool: True if the image was successfully processed and saved, False otherwise.
    """
    try:
        # Read the image
        img = cv2.imread(image_path)

        if img is None:
            print(f"Error: Could not load image from {image_path}")
            return False

        # Add the text to the image
        cv2.putText(img, text_to_add, org, font, font_scale, color, thickness, cv2.LINE_AA)

        # Save the modified image
        cv2.imwrite(output_path, img)

        print(f"Text successfully added to the image and saved to {output_path}")
        return True
    except Exception as e:
        print(f"An error occurred: {e}")
        return False


# --- Example Usage --- #
# First, let's create a dummy image for demonstration purposes
dummy_image = np.zeros((300, 500, 3), dtype=np.uint8)
cv2.putText(dummy_image, "Original Image", (100, 150), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
cv2.imwrite('original_image.jpg', dummy_image)

print("Created a dummy image 'original_image.jpg'.")

# Now, use the function to add text
image_file = 'original_image.jpg'
user_text = "Hello from Colab!"
output_file = 'output_image_with_text.jpg'

success = add_text_to_image(image_file, user_text, output_file, color=(0, 0, 255), org=(80, 200), font_scale=1.2, thickness=3)

if success:
    print(f"You can now find the modified image at: {output_file}")
    # Display the image using cv2_imshow for Colab compatibility
    img_with_text = cv2.imread(output_file)
    if img_with_text is not None:
        cv2_imshow(img_with_text)
    else:
        print("Could not read the output image for display.")
```
