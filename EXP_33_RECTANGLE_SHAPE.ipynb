```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

### Function to Create a White Image and Draw a Rectangle

```python
def create_and_draw_rectangle(rect_x, rect_y, rect_width, rect_height, color=(0, 0, 0), thickness=2, base_image_path=None, img_width=None, img_height=None):
    """
    Creates a white image of specified size or loads an existing image, and draws a rectangle on it.

    Args:
        rect_x (int): The x-coordinate of the top-left corner of the rectangle.
        rect_y (int): The y-coordinate of the top-left corner of the rectangle.
        rect_width (int): The width of the rectangle.
        rect_height (int): The height of the rectangle.
        color (tuple): The color of the rectangle in BGR format (default is black).
        thickness (int): The thickness of the rectangle's border (default is 2). If -1, the rectangle is filled.
        base_image_path (str, optional): Path to an existing image to load. If provided, img_width and img_height are ignored.
        img_width (int, optional): The width of the image to create if base_image_path is not provided. Required if base_image_path is None.
        img_height (int, optional): The height of the image to create if base_image_path is not provided. Required if base_image_path is None.

    Returns:
        numpy.ndarray: The image with the drawn rectangle.
    """
    if base_image_path:
        image = cv2.imread(base_image_path)
        if image is None:
            raise FileNotFoundError(f"Error: Could not load image from {base_image_path}. Please check the path and file integrity.")
    else:
        if img_width is None or img_height is None:
            raise ValueError("If 'base_image_path' is not provided, both 'img_width' and 'img_height' must be specified.")
        # Create a white image (height, width, channels)
        image = 255 * np.ones((img_height, img_width, 3), dtype=np.uint8)

    # Define the top-left and bottom-right corners of the rectangle
    pt1 = (rect_x, rect_y)
    pt2 = (rect_x + rect_width, rect_y + rect_height)

    # Draw the rectangle on the image
    cv2.rectangle(image, pt1, pt2, color, thickness)

    return image
```

### Example Usage

```python
# Path to the user's white image
user_image_path = '/content/White image.avif'

# Define rectangle properties for the user's image
rectangle_x_user = 50
rectangle_y_user = 50
rectangle_width_user = 100
rectangle_height_user = 75
rectangle_color_user = (0, 255, 0) # Green color in BGR
rectangle_thickness_user = 3

# Create the image with the rectangle using the user's image
image_on_user_image = create_and_draw_rectangle(
    rect_x=rectangle_x_user,
    rect_y=rectangle_y_user,
    rect_width=rectangle_width_user,
    rect_height=rectangle_height_user,
    color=rectangle_color_user,
    thickness=rectangle_thickness_user,
    base_image_path=user_image_path
)

# Display the image
print(f"Displaying rectangle on user's image: {user_image_path}")
cv2_imshow(image_on_user_image)

# Keep one of the original examples for creating a new white image
# Define image dimensions for new image
image_width = 600
image_height = 400

# Define rectangle properties for new image
rectangle_x = 100
rectangle_y = 50
rectangle_width = 200
rectangle_height = 150
rectangle_color = (0, 0, 255) # Red color in BGR
rectangle_thickness = 5

# Create a new white image with the rectangle
image_with_rectangle = create_and_draw_rectangle(
    img_width=image_width,
    img_height=image_height,
    rect_x=rectangle_x,
    rect_y=rectangle_y,
    rect_width=rectangle_width,
    rect_height=rectangle_height,
    color=rectangle_color,
    thickness=rectangle_thickness
)

print("\nDisplaying rectangle on a newly created white image:")
cv2_imshow(image_with_rectangle)

# Example with a filled rectangle on a new white image
image_with_filled_rectangle = create_and_draw_rectangle(
    img_width=image_width,
    img_height=image_height,
    rect_x=350,
    rect_y=150,
    rect_width=100,
    rect_height=100,
    color=(255, 0, 0), # Blue color
    thickness=-1
)
print("\nDisplaying filled rectangle on a newly created white image:")
cv2_imshow(image_with_filled_rectangle)
```
