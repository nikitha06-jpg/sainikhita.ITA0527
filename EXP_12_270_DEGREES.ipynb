```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
original_image = Image.open('/content/image 2.jpg')

# Perform a 270-degree clockwise rotation. In Pillow, this is done by passing 270 to the rotate method.
rotated_image = original_image.rotate(270, expand=True)

# Display the original and rotated images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rotated_image)
plt.title('Rotated 270 Degrees Clockwise')
plt.axis('off')

plt.tight_layout()
plt.show()
```
