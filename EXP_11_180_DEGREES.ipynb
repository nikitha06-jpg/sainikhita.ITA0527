```python
import requests
from PIL import Image
from io import BytesIO

# Path to the local image file
image_path = "/content/images.nikki.jpg"

# Open the image using Pillow
original_image = Image.open(image_path)

print("Original Image:")
display(original_image)
```

```python
# Perform a 180-degree rotation
# A 180-degree rotation is equivalent to flipping both horizontally and vertically.
rotated_image = original_image.rotate(180, expand=True)

print("Image after 180-degree rotation:")
display(rotated_image)
```
