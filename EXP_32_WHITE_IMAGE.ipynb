```python
import numpy as np
import matplotlib.pyplot as plt

def create_image_with_corner_boxes(width, height, background_color=(255, 255, 255)):
    """
    Creates an image of the specified size with a given background color and
    adds four colored boxes (Black, Blue, Green, Red) to each corner.

    Args:
        width (int): The width of the image in pixels.
        height (int): The height of the image in pixels.
        background_color (tuple): An RGB tuple (e.g., (255, 255, 255) for white)
                                  specifying the background color.
    """
    if not isinstance(width, int) or not isinstance(height, int) or width <= 0 or height <= 0:
        raise ValueError("Width and height must be positive integers.")
    if not (isinstance(background_color, tuple) and len(background_color) == 3 and
            all(0 <= c <= 255 for c in background_color)):
        raise ValueError("Background color must be an RGB tuple with values between 0 and 255.")

    # Create the image with the specified background color
    image = np.full((height, width, 3), background_color, dtype=np.uint8)

    # Calculate box dimensions (1/10th of image size)
    box_width = width // 10
    box_height = height // 10

    # Define colors in RGB
    black = np.array([0, 0, 0], dtype=np.uint8)
    blue = np.array([0, 0, 255], dtype=np.uint8)
    green = np.array([0, 255, 0], dtype=np.uint8)
    red = np.array([255, 0, 0], dtype=np.uint8)

    # Place boxes in corners
    # Top-left (Black)
    image[0:box_height, 0:box_width] = black
    # Top-right (Blue)
    image[0:box_height, width - box_width:width] = blue
    # Bottom-left (Green)
    image[height - box_height:height, 0:box_width] = green
    # Bottom-right (Red)
    image[height - box_height:height, width - box_width:width] = red

    # Display the image
    plt.imshow(image)
    plt.title(f'Image with Corner Boxes ({width}x{height}) - Background: {background_color}')
    plt.axis('off') # Hide axes
    plt.show()

# Example usage:
create_image_with_corner_boxes(800, 600, background_color=(0, 0, 0)) # Black background
```

You can call the function with different `width` and `height` values to create images of various sizes, for example:


```python
create_image_with_corner_boxes(1000, 400, background_color=(192, 192, 192)) # Light gray background
create_image_with_corner_boxes(500, 500) # Default white background
```
